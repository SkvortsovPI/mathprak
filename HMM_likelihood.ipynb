Задание: для HMM, заданной на картинке ниже, рассчитать правдоподобие последовательности наблюдаемых 13323332121131333. Для расчета реализовать алгоритм прямого прохода (forward algorithm).

```python
import numpy as np

M = 2
P = np.array ([
    [.5, .5],
    [.4, .6]])
obs = np.array ([int(c) - 1 for c in str(13323332121131333)])
pi_0 = np.array ([.2, .8])
A = np.array ([
    [.5, .4, .1],
    [.2, .4, .4]])
```

```python
alpha_curr = np.array ([pi_0[i] * A[i][obs[0]] for i in range(M)]) # t = 1
for i in range (1, len(obs)): # t = i+1 
    alpha_prev, alpha_curr = alpha_curr, np.zeros (M)
    for j in range (M):
        for k in range (M):
            alpha_curr[j] += alpha_prev[k] * P[k][j] * A[j][obs[i]]
print ("likelihood =", np.sum (alpha_curr))
```
